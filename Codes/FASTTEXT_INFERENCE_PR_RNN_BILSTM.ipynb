```python
import numpy as np
import re
import pandas as pd
import pickle
from tensorflow.keras.models import load_model
from gensim.models import FastText
```

```python
# Clean Bengali text
def process_comments(comment):
    return re.sub('[^\u0980-\u09FF]', ' ', str(comment))

# Remove Bengali stop words
def remove_stop_words(text, stop_words):
    if isinstance(text, str):
        words = text.split()
        return ' '.join([word for word in words if word not in stop_words])
    return text



# Convert text to FastText embedding
def text_to_embedding(text, model, max_length=100):
    zero_vector = np.zeros(model.vector_size)
    # words = text.split()
    embedding = np.zeros((max_length, model.vector_size))
    for i, word in enumerate(text[:max_length]):
        embedding[i] = model.wv[word] if word in model.wv else zero_vector
    # for i, word in enumerate(words[:max_length]):
    #     embedding[i] = model.wv[word] if word in model.wv else zero_vector
    return embedding
```

```python
def predict_all_models(sentences, max_length=100):
    # Assuming models and necessary files are loaded globally or passed to the function

    # Step 1: Preprocessing
    cleaned = [process_comments(s) for s in sentences]
    filtered = [remove_stop_words(s, stop_words_list) for s in cleaned]

    # Step 2: Convert to embeddings
    x_emb = np.array([text_to_embedding(text, fasttext_model, max_length) for text in filtered])


    # Step 3: Predict with each model and decode
    print("Predictions:")
    models = {
        "RNN": rnn_model,
        "BiLSTM": bilstm_model,
        "CNN": cnn_model,
        "CNN_BiLSTM": cnn_bilstm_model,
        "BGRU": bgru_model,
        "CNN_BGRU": cnn_bgru_model,
    }

    for i, text in enumerate(sentences):
        print(f"\nğŸ“Œ Input: {text}")
        for model_name, model in models.items():
            try:
                # Get the embedding for the current sentence
                sentence_emb = np.array([x_emb[i]])
                # Predict
                y_pred = model.predict(sentence_emb)
                # Decode prediction
                pred_label = le.inverse_transform([np.argmax(y_pred)])
                print(f"â†’ {model_name} Prediction: {pred_label[0]}")
            except Exception as e:
                print(f"Error predicting with {model_name}: {e}")
```

```python
from gtts import gTTS
import os
import numpy as np
import re
import pandas as pd
import pickle
from tensorflow.keras.models import load_model
from gensim.models import FastText

# Load models (ensure correct path if needed)
# Assuming models are in a 'models' directory in the content folder after gdown
try:
    rnn_model = load_model('/content/models/rnn_model.h5')
    bilstm_model = load_model('/content/models/bilstm_model.h5')
    cnn_model = load_model('/content/models/cnn_model.h5')
    cnn_bilstm_model = load_model('/content/models/cnn_bilstm_model.h5')
    bgru_model = load_model('/content/models/bgru_model.h5')
    cnn_bgru_model = load_model('/content/models/cnn_bgru_model.h5')

    # Load FastText model
    fasttext_model = FastText.load("/content/models/fasttext_model/fasttext_model.model")

    # Load LabelEncoder
    with open("/content/models/label_encoder.pkl", "rb") as f:
        le = pickle.load(f)

    # Load stop words (ensure correct path if needed)
    # Assuming stopwords_bangla.csv is in the content folder
    stop_words_df = pd.read_csv("/content/models/stopwords_bangla.csv").drop(['Unnamed: 0'], axis=1)
    stop_words_list = stop_words_df.iloc[:, 0].tolist()

    print("All models and necessary files loaded successfully.")

except FileNotFoundError as e:
    print(f"Error loading files: {e}. Please ensure all necessary files are in the /content/models/ directory.")
except Exception as e:
    print(f"An unexpected error occurred while loading files: {e}")

# Clean Bengali text
def process_comments(comment):
    return re.sub('[^\u0980-\u09FF]', ' ', str(comment))

# Remove Bengali stop words
def remove_stop_words(text, stop_words):
    if isinstance(text, str):
        words = text.split()
        return ' '.join([word for word in words if word not in stop_words])
    return text



# Convert text to FastText embedding
def text_to_embedding(text, model, max_length=100):
    zero_vector = np.zeros(model.vector_size)
    # words = text.split()
    embedding = np.zeros((max_length, model.vector_size))
    for i, word in enumerate(text[:max_length]):
        embedding[i] = model.wv[word] if word in model.wv else zero_vector
    # for i, word in enumerate(words[:max_length]):
    #     embedding[i] = model.wv[word] if word in model.wv else zero_vector
    return embedding


def predict_emotion(text, model_name, max_length=100):
    # Step 1: Preprocessing
    cleaned_text = process_comments(text)
    filtered_text = remove_stop_words(cleaned_text, stop_words_list)

    # Step 2: Convert to embeddings
    x_emb = np.array([text_to_embedding(filtered_text, fasttext_model, max_length)])

    # Step 3: Predict with the selected model
    if model_name == "RNN":
        model = rnn_model
    elif model_name == "BiLSTM":
        model = bilstm_model
    elif model_name == "CNN":
        model = cnn_model
    elif model_name == "CNN_BiLSTM":
        model = cnn_bilstm_model
    elif model_name == "BGRU":
        model = bgru_model
    elif model_name == "CNN_BGRU":
        model = cnn_bgru_model
    else:
        return "Invalid model name selected."

    # Predict
    y_pred = model.predict(x_emb)
    # Decode prediction
    label = le.inverse_transform([np.argmax(y_pred)])[0]


    # Step 4: Generate audio
    tts = gTTS(text=text, lang='bn')
    audio_file = "output.mp3"
    tts.save(audio_file)

    return label, audio_file
```

```python
import gradio as gr

# Define the Gradio interface
interface = gr.Interface(
    fn=predict_emotion,
    inputs=[
        gr.Textbox(label="Enter Bengali text"),
        gr.Dropdown(
            ["RNN", "BiLSTM", "CNN", "CNN_BiLSTM", "BGRU", "CNN_BGRU"],
            label="Select Model",
            value="RNN" # Default value
        )
    ],
    outputs=[
        gr.Textbox(label="Predicted Emotion"),
        gr.Audio(label="Input Text Audio")
    ],
    title="Bengali Emotion Detection",
    description="Enter Bengali text and select a model to predict the emotion and hear the text."
)

# Launch the interface
interface.launch(debug=True)
```
